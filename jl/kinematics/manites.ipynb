```julia
using Plots

function logistic(a, num)
    u0 = 0.01
    u = u0
    u_sq = [ u0 ]
    
    for i in 1:num
        u = a*u*(1-u)
        push!(u_sq, u)
    end
    
    return u_sq
end
```

```julia
fig = plot( logistic(1.2, 50), marker = :circle )
```

```julia
plot(fig, xlabel='X', ylabel='Y', xlims=(0,100),)
```
